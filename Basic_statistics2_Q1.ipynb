```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
companies=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers','Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways','Warner-Lambert']
```

```python
measures = [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00]
```

```python
plt.figure(figsize=(12,6))
plt.bar(companies,measures)
plt.xticks(rotation=90)
plt.xlabel('companies')
plt.ylabel('measures')
plt.title('X MEASURES OF COMPANIES')
plt.show()
```

```python
import statistics
```

```python
statistics.mean(measures)
```

```python
statistics.variance(measures)
```

```python
measures_array=np.array(measures)
print(measures_array)
```

```python
mean_measures=np.mean(measures_array)
print(mean_measures)
```

```python
var_measures=np.var(measures_array)
print(var_measures)
```

```python
std_dv=np.std(measures_array)
print(std_dv)
```

```python
import pandas as pd
```

```python
x_measures=[24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00]
```

```python
df=pd.DataFrame(x_measures)
```

```python
#calculate range of Q3 and Q1 
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
```

```python
lower_bound=Q1-IQR*1.5
upper_bound=Q3+IQR*1.5
```

```python
outliers = df < (lower_bound)
print(outliers)
```

```python
outliers=df >(upper_bound)
print(outliers)
```
