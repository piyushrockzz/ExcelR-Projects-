## Data Preprocessing

```python
df.head()
```

```python
len(df)
```

```python
df['quarter'] = 0
for i in range(42):
    p=df['Quarter'][i]
    df['quarter'][i]=p[0:2]
```

```python
df
```

```python
df['quarter'].value_counts()
```

```python
df_dummies=pd.DataFrame(pd.get_dummies(df['quarter']),columns=['Q1','Q2','Q3','Q4'])
cc=pd.concat([df,df_dummies],axis= 1)
```

```python
df.head()
```

```python
cc
```

```python
cc['t'] = np.arange(1,43)
cc['t_squared'] = cc['t']**2
cc["Sales_log"] =np.log(df['Sales'])
```

```python
cc.head()
```

```python
train =cc.head(32)
test =cc.tail(10)
```

```python
df['Sales'].plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Visualize the data
plt.figure(figsize=(10, 6))
plt.plot(df['Quarter'], df['Sales'])
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Coca-Cola Sales Data')
plt.show()

# Calculate Rolling Statistics
rolling_mean = df['Sales'].rolling(window=4).mean()
rolling_std = df['Sales'].rolling(window=4).std()

plt.figure(figsize=(10, 6))
plt.plot(df['Quarter'], df['Sales'], label='Original')
plt.plot(df['Quarter'], rolling_mean, label='Rolling Mean (window=4)')
plt.plot(df['Quarter'], rolling_std, label='Rolling Std (window=4)')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Rolling Mean and Standard Deviation')
plt.legend()
plt.show()

# Perform Augmented Dickey-Fuller Test
result = adfuller(df['Sales'])

print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value}')
```

### Models

```python
from sklearn.metrics import mean_squared_error
```

```python
# Linear Model
linear_model =smf.ols("Sales~t",data =train).fit()
linear_pred = pd.Series(linear_model.predict(test['t']))
linear_rmse =np.sqrt(mean_squared_error(np.array(test['Sales']),np.array(linear_pred)))
linear_rmse
```

```python
#Quadratic Model
quad_model =smf.ols("Sales~t+t_squared",data=train).fit()
quad_pred = pd.Series(quad_model.predict(test[['t','t_squared']]))
quad_rmse =np.sqrt(mean_squared_error(np.array(test['Sales']),np.array(quad_pred)))
quad_rmse
```

```python
# Exponential model
exp_model  =smf.ols("Sales_log~t",data=train).fit()
exp_pred =pd.Series(exp_model.predict(test['t']))
exp_rmse =np.sqrt(mean_squared_error(np.array(test['Sales']),np.array(exp_pred)))
exp_rmse
```

```python
data = {"MODEL":pd.Series(["rmse_linear","rmse_exp","rmse_quad"]),"RMSE_Values":pd.Series([linear_rmse,exp_rmse,quad_rmse,])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```

```python
from pandas import read_csv
from sklearn.metrics import mean_squared_error
from math import sqrt

train = read_csv('dataset_cc.csv', header=None, index_col=0, parse_dates=True, squeeze=True)

train
```

```python
X = train.values
X = X.astype('float32')
train_size = int(len(X) * 0.50)
train, test = X[0:train_size], X[train_size:]
```

###  Validation

```python
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    yhat = history[-1]
    predictions.append(yhat)
# observation
    obs = test[i]
    history.append(obs)
    print('>Predicted=%.3f, Expected=%.3f' % (yhat, obs))
# report performance
rmse = sqrt(mean_squared_error(test, predictions))
print('RMSE: %.3f' % rmse)
```

```python
data = {"MODEL":pd.Series(["rmse_linear","rmse_exp","rmse_quad","RMSE_ARIMA"]),"RMSE_Values":pd.Series([linear_rmse,exp_rmse,quad_rmse,rmse])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```
