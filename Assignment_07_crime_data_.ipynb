```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

data = pd.read_csv('crime_data.csv', index_col=0)
```

```python
# Perform hierarchical clustering
linkage_matrix = linkage(data, method='complete', metric='euclidean')
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, labels=data.index, orientation='left')
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Distance')
plt.ylabel('States')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)


```

```python
# Get the cluster labels
cluster_labels = kmeans.labels_
```

```python
# Get the centroids of each cluster
centroids = kmeans.cluster_centers_
```

```python

# Print the number of clusters formed
num_clusters = len(set(cluster_labels))
print("Number of clusters formed:", num_clusters)
```

```python
from sklearn.cluster import DBSCAN
```

```python
# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.3, min_samples=2)
dbscan.fit(data)
```

```python
# Get the cluster labels
cluster_labels = dbscan.labels_
```

```python

# Get the number of clusters formed (excluding noise points)
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
print("Number of clusters formed:", num_clusters)
```
